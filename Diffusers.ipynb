```python

from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id = "dreamlike-art/dreamlike-diffusion-1.0"
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
```

```python
pipe.config
```

```python
pipe = pipe.to("cuda")
```

```python
images = pipe(
    prompt = "Railway track Snowplow. A digital picture. The best in the artsation. It stands on the railway against the background of a field",
    height = 1024,
    width = 512,
    num_inference_steps = 100,
    guidance_scale = 8.5,
    num_images_per_prompt = 1
).images
```
